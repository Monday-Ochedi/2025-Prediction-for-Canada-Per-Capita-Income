```python
# Import Neccesary Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# Load Dataset

df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
# Plot Scatter Plot to Show Points

plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'],df['per capita income (US$)'], color = 'red', marker = '+')
```

```python
# Create a Linear Regression Object

reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
# Check for Coefficent

reg.coef_
```

```python
# Check for Intercept

reg.intercept_
```

```python
# Plot Scatter Plot with Regression Line

plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'],df['per capita income (US$)'], color = 'red', marker = '+')
plt.plot(df['year'], reg.predict(df[['year']]), color = 'blue')
```

```python
# Ignore any warning that may pop up

import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
# Make Prediction

reg.predict([[2025]])
```

```python
# Add my Prediction to my Actual Scatter Plot 

plt.figure(figsize=(10, 5))
plt.gca().set_facecolor('lightgrey')
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'],df['per capita income (US$)'], color = 'red', marker = '+')
plt.plot(df['year'], reg.predict(df[['year']]), color = 'blue')
plt.scatter(2025, reg.predict([[2025]]), color='purple', marker='o', label='Prediction (2025)')
plt.title('2025 Prediction for Canada Per Capita Income')
plt.legend()
```
